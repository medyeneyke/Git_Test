```python
# 2. Visualization Tasks

# Plot 1: Bar chart for "Country" vs. "Population (Millions)"
plt.figure(figsize=(12, 7))
plt.bar(df['Country'], df['Population (Millions)'], color='skyblue')
plt.xlabel("Country")
plt.ylabel("Population (Millions)")
plt.title("Population by Country")
plt.xticks(rotation=90)
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# Plot 2: Scatter plot for "GDP (Trillions)" vs. "Life Expectancy (Years)"
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP (Trillions)'], df['Life Expectancy (Years)'], color='green', alpha=0.7)
plt.xlabel("GDP (Trillions)")
plt.ylabel("Life Expectancy (Years)")
plt.title("GDP vs. Life Expectancy")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot 4: Pie chart for "Renewable Energy (%)"
# Showing the top 5 countries for better pie chart readability
df_sorted_renewable = df.sort_values('Renewable Energy (%)', ascending=False).head(5)
plt.figure(figsize=(8, 8))
plt.pie(df_sorted_renewable['Renewable Energy (%)'], labels=df_sorted_renewable['Country'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("Top 5 Countries by Renewable Energy (%)")
plt.axis('equal') # Ensures pie is drawn as a circle
plt.tight_layout()
plt.show()
```

```python
# Plot 6: Box plot for "Internet Users (%)"
plt.figure(figsize=(8, 6))
# df['Internet Users (%)'].dropna() can be used if missing values were possible and not handled
plt.boxplot(df['Internet Users (%)'])
plt.ylabel("Internet Users (%)")
plt.title("Box Plot of Internet Users (%)")
plt.xticks([1], ['Overall Distribution']) # Label the x-axis tick for clarity
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot 7: Heatmap showing the correlation between numeric variables
plt.figure(figsize=(12, 10)) # Increased figure size for better readability
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5) # Added linewidths
plt.title("Correlation Heatmap of Numeric Variables")
plt.tight_layout()
plt.show()
```

```python

# Plot 8: Stacked bar chart for "Population (Millions)" and "CO2 Emissions (MT)" by country.
# A grouped bar chart is generally more appropriate for comparing two different scales.
# If strictly stacked is needed, normalization or dual-axis might be considered, but can be misleading.
# Let's provide a grouped bar chart.

fig, ax1 = plt.subplots(figsize=(16, 9)) # Increased figure size

bar_width = 0.35
index = range(len(df['Country']))

# Population bars on primary y-axis
bar1 = ax1.bar(index, df['Population (Millions)'], bar_width, label='Population (Millions)', color='dodgerblue')

# Create a second y-axis for CO2 Emissions
ax2 = ax1.twinx()
bar2 = ax2.bar([i + bar_width for i in index], df['CO2 Emissions (MT)'], bar_width, label='CO2 Emissions (MT)', color='crimson')

ax1.set_xlabel('Country', fontweight='bold')
ax1.set_ylabel('Population (Millions)', color='dodgerblue', fontweight='bold')
ax2.set_ylabel('CO2 Emissions (MT)', color='crimson', fontweight='bold')
ax1.set_title('Population and CO2 Emissions by Country (Grouped)', fontweight='bold', fontsize=16)
ax1.set_xticks([i + bar_width / 2 for i in index])
ax1.set_xticklabels(df['Country'], rotation=90, ha="right")

# Add legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# Plot 9: Area chart for "GDP (Trillions)"
# Sorting by GDP can make the area chart more visually coherent for comparison.
df_sorted_gdp = df.sort_values('GDP (Trillions)')
plt.figure(figsize=(12, 6))
plt.fill_between(df_sorted_gdp['Country'], df_sorted_gdp['GDP (Trillions)'], color="skyblue", alpha=0.5)
plt.plot(df_sorted_gdp['Country'], df_sorted_gdp['GDP (Trillions)'], color="Slateblue", alpha=0.8, linewidth=2)
plt.xlabel("Country (Sorted by GDP)")
plt.ylabel("GDP (Trillions)")
plt.title("Area Chart of GDP by Country")
plt.xticks(rotation=90)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Plot 10: Custom visualization of your choice.
# Bubble chart: "Happiness Index" vs "Life Expectancy (Years)",
# with bubble size representing "Population (Millions)" and color representing "GDP (Trillions)".
plt.figure(figsize=(14, 9))

# Normalize population for better bubble size scaling
pop_min = df['Population (Millions)'].min()
pop_max = df['Population (Millions)'].max()
# Avoid division by zero if all populations are the same (unlikely here)
if pop_max - pop_min == 0:
    pop_normalized = [1] * len(df) # All bubbles same size
else:
    pop_normalized = (df['Population (Millions)'] - pop_min) / (pop_max - pop_min)

bubble_sizes = pop_normalized * 2000 + 50  # Adjust scaling factor and minimum size for visibility

scatter_plot = plt.scatter(
    df['Happiness Index'],
    df['Life Expectancy (Years)'],
    s=bubble_sizes,
    c=df['GDP (Trillions)'],
    alpha=0.7,
    cmap='viridis', # Colormap for GDP
    edgecolors='black', # Add edge color to bubbles
    linewidths=0.5
)

plt.xlabel("Happiness Index", fontweight='bold')
plt.ylabel("Life Expectancy (Years)", fontweight='bold')
plt.title("Custom Bubble Chart: Happiness vs. Life Expectancy", fontweight='bold', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.5)

# Add a colorbar for GDP
cbar = plt.colorbar(scatter_plot, label='GDP (Trillions)')
cbar.set_label('GDP (Trillions)', fontweight='bold')

# Optionally, add text labels for some countries (can be crowded)
# for i, txt in enumerate(df['Country']):
#    if df['Population (Millions)'][i] > 200 or df['GDP (Trillions)'][i] > 10: # Example condition
#        plt.annotate(txt, (df['Happiness Index'][i], df['Life Expectancy (Years)'][i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.tight_layout()
plt.show()
```
